# Fluorescence profile analysis of regions of interest from different time-lapse videos in microfluidic devices of different yeast co-cultures.

```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

The excell files contain the fluorescence profile of each vertical line across the full width of the ROI of all slices of the video, one column per slice of the video/time step.

## MCR5101 and HKE3402

```python
# Excell with measurements
path = '/content/drive/MyDrive/Marta/Microfluidics/mCherry analysis/Results_Control_moviePos6.csv'

# read the data file and create dataframe
df = pd.read_csv(path)
# x axis (width of ROI)
x_cp6 = np.asarray(df['x'])
# Get values in time n
time_60_cp6 = np.asarray(df['60'])
time_100_cp6 = np.asarray(df['100'])
time_200_cp6 = np.asarray(df['200'])
```

```python
# Excell with measurements
path = '/content/drive/MyDrive/Marta/Microfluidics/mCherry analysis/Results_Control_moviePos5.csv'

# read the data file and create dataframe
df = pd.read_csv(path)
# x axis (width of ROI)
x_cp5 = np.asarray(df['x'])
# Get values in time n
time_60_cp5 = np.asarray(df['60'])
time_100_cp5 = np.asarray(df['100'])
time_200_cp5 = np.asarray(df['200'])
```

## FLO11- MCR5101

This co-culture has no mCherry expression, in this case the analysis is performed over the GFP fluorescence signal (Channel 001).

```python
# Excell with measurements
path = '/content/drive/MyDrive/Marta/Microfluidics/mCherry analysis/ResultsPos17Ch001.csv'

# read the data file and create dataframe
df = pd.read_csv(path)
# x axis (width of ROI)
x_fm17 = np.asarray(df['x'])
# Get values in time n
time_60_fm17 = np.asarray(df['60'])
time_150_fm17 = np.asarray(df['150'])
```

```python
# Excell with measurements
path = '/content/drive/MyDrive/Marta/Microfluidics/mCherry analysis/ResultsPos35Ch001.csv'

# read the data file and create dataframe
df = pd.read_csv(path)
# x axis (width of ROI)
x_fm35 = np.asarray(df['x'])
# Get values in time n
time_60_fm35 = np.asarray(df['60'])
time_150_fm35 = np.asarray(df['150'])
```

## Create a combined plot with the results

```python
plt.style.use('seaborn')

fig = plt.figure(figsize=(17, 18), layout='constrained')

subfigs = fig.subfigures(nrows=3, ncols=1, wspace=0.7, hspace=5)

ax = subfigs[0].subplots(nrows=1, ncols=2)

title = subfigs[0].suptitle('MCR5101 and HKE3402',  y = 1.0, fontsize = 20, backgroundcolor='#dedad2')
title._bbox_patch._mutation_aspect = 0.02
title.get_bbox_patch().set_boxstyle("square", pad=15)

ax[0].plot(x_cp6, time_200_cp6, color = 'red', linestyle='dashed', label = 'Time 200 (34 hr)')
ax[0].plot(x_cp6, time_60_cp6, color = 'red', label = 'Time 60 (10 hr)')
ax[0].set_title('Pos 6')
ax[0].set_ylabel('mCherry intensity', fontsize = 25)

ax[1].plot(x_cp5, time_200_cp5, color = 'red', linestyle='dashed', label = 'Time 200 (34 hr)')
ax[1].plot(x_cp5, time_60_cp5, color = 'red', label = 'Time 60 (10 hr)')
ax[1].set_title('Pos 5')
ax[1].legend(loc='upper left', bbox_to_anchor=(1, 0.5), fontsize = 20)

ax2 = subfigs[1].subplots(nrows=1, ncols=2)

title2 = subfigs[1].suptitle('FLO11 - rep and HKE3402',  y = 1.0, fontsize = 20, backgroundcolor = '#dedad2')
title2._bbox_patch._mutation_aspect = 0.02
title2.get_bbox_patch().set_boxstyle("square", pad=15)

ax2[0].plot(x_fhp8, time_200_fhp8, color = 'purple', linestyle='dashed', label = 'Time 200 (34 hr)')
ax2[0].plot(x_fhp8, time_60_fhp8, color = 'purple', label = 'Time 60 (10 hr)')
ax2[0].set_title('Pos 8')
ax2[0].set_ylabel('mCherry intensity', fontsize = 25)

ax2[1].plot(x_fhp4, time_200_fhp4, color = 'purple', linestyle='dashed', label = 'Time 200 (34 hr)')
ax2[1].plot(x_fhp4, time_60_fhp4, color = 'purple', label = 'Time 60 (10 hr)')
ax2[1].set_title('Pos 4')
ax2[1].legend(loc='upper left', bbox_to_anchor=(1, 0.5), fontsize = 20)


ax3 = subfigs[2].subplots(nrows=1, ncols=2)

title3 = subfigs[2].suptitle('FLO11 - and MCR5101',  y = 1.0, fontsize = 20, backgroundcolor = '#dedad2')
title3._bbox_patch._mutation_aspect = 0.02
title3.get_bbox_patch().set_boxstyle("square", pad=15)

ax3[0].plot(x_fm35, time_150_fm35, color = 'green', linestyle='dashed',label = 'Time 150 (25h)')
ax3[0].plot(x_fm35, time_60_fm35, color = 'green',label = 'Time 60 (10h)')
ax3[0].set_title('Pos 35')
ax3[0].set_xlabel('ROI width', fontsize = 20)
ax3[0].set_ylabel('GFP intensity', fontsize = 25)

ax3[1].plot(x_fm17, time_150_fm17, color = 'green', linestyle='dashed', label = 'Time 150 (25h)')
ax3[1].plot(x_fm17, time_60_fm17, color = 'green', label = 'Time 60 (10h)')
ax3[1].set_title('Pos 17')
ax3[1].set_xlabel('ROI width', fontsize = 20)
ax3[1].legend(loc='upper left', bbox_to_anchor=(1, 0.5), fontsize = 20)


fig.suptitle('Intensity along ROI width', y = 1.04, fontsize = 30)
plt.subplots_adjust(top=0.8)
plt.show()
```
